````python
def _summarize_meeting_transcript_chunk(meeting_transcript_chunk):
    meeting_chunk_summarization_prompt_template = \
    """<|system|>
    You are a smart AI Large Language Model skilled in summarizing text. You are able to read large amounts of texts, extract the most important vital key insights.
    Your task is to summarize the given part of meeting transcript (i.e. it is incomplete) in a consise and informative manner.
    Give the summary in bullet points form with dash (-) to represent each point. You should not include any irrelevant information or details that are not important to the main topic of the meeting. You should also not include any personal opinions or comments in your summary. Your summary should be accurate, informative, and easy to understand.
    Make sure the summary does not exceed 2000 words.

    <|user|>
    Summarize the following part of the meeting transcript, delimitted by tripple single quotes:
    '''
    {transcript}
    '''

    <|assistant|>
    """
    generated_response = model.generate(prompt=meeting_chunk_summarization_prompt_template.format(transcript=meeting_transcript_chunk))
    
    # parse the response
    summary = generated_response['results'][0]['generated_text']
    summary = summary[summary.index('<|assistant|>') + len('<|assistant|>'):]
    summary = summary.strip()
    
    return summary

def _summarize_meeting_transcript(meeting_transcript, is_summarized_chunks=False):
    
    if not is_summarized_chunks:
        meeting_summarization_prompt_template = \
        """<|system|>
        You are a smart AI Large Language Model skilled in summarizing text. You are able to read large amounts of texts, extract the most important vital key insights, and present them in an easy to read format.
        Your task is to summarize the given meeting transcript in a consise and informative manner. The goal is to have a summary that can serve as a quick review of the meeting for those who attended or as a reference for those who missed the meeting, as such it shouldn't be long.
        You should first read and understand the meeting transcript text carefully and then extract the key points, most important details, and main ideas from the transcript and present them in a clear and organized way. Do not include any irrelevant information, details that are not present in the meeting, or any personal opinions or comments in your summary. 
        Give your response in markdown format with headings and bullet points to organize the information. Start off by giving a 1 sentence brief overview of what the meeting was about followed by the key points and main ideas discussed during the meeting.
        Format your reply as follows and don't say anything else:

        ------------------------
        SUMMARY
        ```
        # Overview
        <One sentence overview of the meeting>

        # <Key Topic 1>
        - <Key point 1>
        - <Key point 2>
        ...

        # <Key Topic 2>
        - <Key point 1>
        - <Key point 2>
        ...

        ...
        ...
        ```
        
        <|user|>
        Summarize the following meeting transcript, delimitted by tripple single quotes:
        '''
        {transcript}
        '''

        <|assistant|>
        Summary in markdown format:
        """

    else:
        meeting_summarization_prompt_template = \
        """<|system|>
        You are a smart AI Large Language Model skilled in summarizing text. You are able to read large amounts of texts, extract the most important vital key insights, and present them in an easy to read format.
        Your task is to summarize the given meeting transcript sumaries in a consise and informative manner, that is your input will be a bunch of summaries of different chunks of the meeting transcript.
        The goal is to have a summary that can serve as a quick review of the meeting for those who attended or as a reference for those who missed the meeting, as such it shouldn't be long.
        You should first read and understand all the individual meeting summary texts carefully and then extract the key points, most important details, and main ideas from them all and collect, organize and present them in a clear way. Do not include any irrelevant information, details that are not present in the meeting, or any personal opinions or comments in your summary. 
        Give your response in markdown format with headings and bullet points to organize the information. Start off by giving a 1 sentence brief overview of what the meeting was about followed by the key points and main ideas discussed during the meeting.
        Format your reply as follows and don't say anything else:
        
        ------------------------
        SUMMARY
        ```
        # Overview
        <One sentence overview of the meeting>

        # <Key Topic 1>
        - <Key point 1>
        - <Key point 2>
        ...

        # <Key Topic 2>
        - <Key point 1>
        - <Key point 2>
        ...

        ...
        ...
        ```
        
        <|user|>
        Summarize the following meeting transcript, delimitted by tripple single quotes:
        '''
        {transcript}
        '''

        <|assistant|>
        Summary in markdown format:
        """
    
    generated_response = model.generate(prompt=meeting_summarization_prompt_template.format(transcript=meeting_transcript))

    # parse the response
    summary = generated_response['results'][0]['generated_text']
    summary = summary[summary.index('<|assistant|>') + len('<|assistant|>'):]
    summary = summary.strip()

    return summary

def get_transcripts():
    transcripts_file = "./Data/transcripts.json"
    with open(transcripts_file, "r") as f:
        transcripts = json.load(f)['transcripts']
    return transcripts

def summarize_meeting_transcript(transcript):
    text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
        chunk_size=3000,
        chunk_overlap=20,
    )
    texts = text_splitter.split_text(transcript)
    if len(texts) == 1:
        print("Transcript is small enough to summarize in one go.")
        return _summarize_meeting_transcript(transcript, is_summarized_chunks=False)
    else:
        print(f"Transcript is too large to summarize in one go. Splitting into {len(texts)} chunks.")
        summaries = []
        for i, chunk in enumerate(texts):
            print(f"Summarizing chunk {i + 1}/{len(texts)}")
            chunk_summary = _summarize_meeting_transcript_chunk(chunk)
            print(chunk_summary, "\n\n")
            summaries.append(f"Summary for chunk {i + 1}/{len(texts)} of the meeting transcript:\n{chunk_summary}")
        return _summarize_meeting_transcript("\n".join(summaries), is_summarized_chunks=True)

````
